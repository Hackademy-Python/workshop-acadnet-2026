## 1. Introducere, Tipuri de date și reprezentarea în memorie

### Descriere concept
Python este un limbaj **interpretat**: codul este executat instrucțiune cu instrucțiune de către interpretor.

Pentru un începător, asta înseamnă:
- **feedback rapid**: scrii câteva linii și vezi imediat rezultatul;
- **iterare ușoară**: poți testa idei mici fără pași complicați de build.

În Python, variabilele sunt etichete către obiecte din memorie, iar tipul este asociat valorii, nu numelui variabilei.

Tipuri de date de bază:
- **`int`** pentru numere întregi;
- **`float`** pentru numere reale;
- **`str`** pentru text;
- **`bool`** pentru valori logice (`True`/`False`).

### Diferențe față de C++
În C++, de obicei declari tipul explicit (`int x = 10;`).

În Python scrii doar `x = 10`, iar tipul este dedus automat.

- C++ este **compilat** (compilare înainte de execuție);
- Python este **interpretat** (rulează direct, foarte bun pentru testare rapidă).

### Cod și explicații
În celula de cod următoare vezi tipare dinamică, conversii de tip și cum verificăm identitatea unui obiect din memorie cu `id()`.


```python
# 1) Tipuri de date + memorie
valoare = 42
print("valoare:", valoare, "| tip:", type(valoare))

valoare = "42"
print("valoare:", valoare, "| tip:", type(valoare))

numar_real = float(valoare)
print("float(valoare):", numar_real, "| tip:", type(numar_real))

numar_int = int(numar_real)
print("int(numar_real):", numar_int, "| tip:", type(numar_int))

a = [1, 2, 3]
b = a
print("id(a):", id(a))
print("id(b):", id(b))
```

### `None` – valoarea absentă

Python are un tip special, `None`, care înseamnă "nicio valoare" sau "necunoscut".
Este echivalentul `nullptr`/`NULL` din C++, dar în Python este folosit mult mai larg:
- o funcție care nu returnează explicit nimic returnează implicit `None`;
- este folosit ca valoare inițială pentru variabile opționale;
- `dict.get(cheie)` returnează `None` dacă cheia nu există (în loc să arunce o eroare).

```python
# None în practică
x = None
print("x este:", x, "| tip:", type(x))

def fara_return():
    pass  # funcție care nu returnează nimic

rezultat = fara_return()
print("Rezultat fara_return():", rezultat)  # None

# None vs 0 vs '' – toate sunt 'falsy', dar sunt valori diferite!
print("None == 0 ?", None == 0)  # False
print("None == '' ?", None == '')  # False

# Mod corect de a verifica None:
valoare = None
if valoare is None:
    print("Variabila nu are o valoare setată încă.")
```

## 2. Operatori (Aritmetici, logici, comparație)

### Descriere concept
Operatorii sunt simboluri cu care facem calcule sau verificări.

În Python avem:
- operatori **aritmetici** (`+`, `-`, `*`, `/`, `//`, `%`, `**`);
- operatori de **comparație** (`==`, `!=`, `<`, `>`, `<=`, `>=`);
- operatori **logici** (`and`, `or`, `not`).

Rezultatul comparațiilor și expresiilor logice este de tip **`bool`** (`True` sau `False`).

### Diferențe față de C++
Multe simboluri sunt identice, dar în Python:
- împărțirea `/` produce mereu **`float`**;
- împărțirea întreagă are operator separat: **`//`**;
- pentru putere folosim direct **`**`**.

### Cod și explicații
Celula următoare arată exemple de operatori și ce rezultate produc.

```python
# 2) Operatori
a, b = 10, 3

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)

print("a > b ->", a > b)
print("a == b ->", a == b)

este_mare = a > 5 and b < 5
print("(a > 5 and b < 5) ->", este_mare)
print("not(a == b) ->", not (a == b))
```

## 3. Operații pe string-uri și combinarea cu numere

### Descriere concept
Un **string** este un text între ghilimele.

Operații uzuale:
- **concatenare** cu `+` (pentru a lipi două texte);
- **repetare** cu `*`;
- extragere de bucăți cu **slicing**;
- verificare sub-șir cu `in`.

**Combinarea textului cu numere:**
Spre deosebire de alte limbaje, în Python **NU poți** aduna direct un text cu un număr (ex: `"Nota mea este " + 10` va genera o eroare de tip `TypeError`). Python are "strong typing" și refuză să ghicească cum vrei să le combini.

Pentru a rezolva asta, ai două opțiuni principale:
1. **Conversia explicită:** Transformi numărul în text folosind funcția `str()`, apoi le concatenezi cu `+`.
2. **f-strings (Format strings):** Aceasta este metoda modernă și recomandată! Pui litera `f` în fața ghilimelelor, iar variabilele de orice tip pot fi introduse direct în text, puse între acolade `{}`.

### Diferențe față de C++
În C++, lucrul cu string-uri implică frecvent `std::string` și metode specifice. Acolo, la afișare, `cout << "Nota: " << 10;` combină automat tipurile pe ecran.
În Python, pentru că nu există o conversie automată a numărului în string la concatenare cu `+`, f-strings (`f"Nota: {10}"`) reprezintă echivalentul cel mai curat și rapid.

### Cod și explicații
Celula de cod arată concatenarea, slicing-ul, dar și cum (nu) putem combina corect numerele cu textul.

```python
# 3) String-uri
prenume = "Maria"
salut = "Salut, " + prenume + "!"
print(salut)

cuvant = "Python"
print("Repetat de 3 ori:", cuvant * 3)

text = "Invatam Python la workshop"
print("Primele 7 caractere:", text[:7])
print("Ultimele 8 caractere:", text[-8:])

print("'Python' apare în text?", "Python" in text)

print("\n--- Combinarea cu numere ---")

nota = 10
materie = "Matematica"

# AȘA NU:
# mesaj_gresit = "Am luat " + nota + " la " + materie 
# Dacă decomentezi linia de mai sus, vei primi TypeError: can only concatenate str (not "int") to str

# AȘA DA (Varianta 1: conversie explicita cu str):
mesaj_clasic = "Am luat " + str(nota) + " la " + materie + "."
print("Varianta cu str():", mesaj_clasic)

# AȘA DA (Varianta 2: f-strings - Metoda modernă și RECOMANDATĂ):
mesaj_fstring = f"Am luat {nota} la {materie}."
print("Varianta cu f-string:", mesaj_fstring)

# f-strings permit și operații matematice direct în acolade!
print(f"Peste un an voi lua {nota + 1}!")
```

## 4. Instrucțiuni de control (If, for, while - folosind indentarea)

### Descriere concept
Instrucțiunile de control decid ce cod se execută și de câte ori.

În Python, blocurile de cod se delimitează prin **indentare** (spații la începutul liniei), nu cu acolade.

Structuri principale:

### Diferențe față de C++
În C++ folosim de obicei acolade `{}` pentru blocuri.

În Python, indentarea corectă este obligatorie și contribuie la un cod mai curat și mai ușor de citit.

### Cod și explicații
Exemplele includ decizie simplă, `for` cu `range` și `while` cu contor.

```python
# 4) Control flow + indentare
nota = 9

if nota >= 9:
    print("Foarte bine")
elif nota >= 7:
    print("Bine")
else:
    print("Trebuie mai mult exercițiu")

print("for cu range(1, 6):")
for i in range(1, 6):
    print(i, end=" ")
print()

print("while până la 3:")
contor = 1
while contor <= 3:
    print("contor =", contor)
    contor += 1
```

## 7. Funcții (inclusiv parametri cheie-valoare) și List Comprehension

### Descriere concept
Funcțiile organizează codul în bucăți reutilizabile.

Ele pot:
- primi parametri;
- returna rezultate;
- fi apelate cu argumente prin **cheie-valoare** (ex: `nume="Ana"`).

**List comprehension** este o metodă scurtă de a crea liste noi din liste existente.

### Diferențe față de C++
În C++, funcțiile folosesc de regulă parametri poziționali și tipuri declarate explicit.

În Python, sintaxa este mai concisă și suportă natural argumente cheie-valoare.

### Cod și explicații
Exemplele includ funcție simplă, apel cu keyword arguments și list comprehension cu filtru.

```python
# 7) Funcții + keyword arguments + list comprehension
def salut(nume, semn="!"):
    return f"Salut, {nume}{semn}"

print(salut("Mara"))
print(salut(nume="Ioan", semn="."))

print("\n--- List Comprehension Simplu ---")
numere = [1, 2, 3, 4, 5, 6]
patrate_pare = [x * x for x in numere if x % 2 == 0]
print("Pătratele numerelor pare:", patrate_pare)

print("\n--- Crearea Matricilor (Listă de liste) ---")
linii = 3
coloane = 4

# Varianta clasică (Procedurală) pentru a crea o matrice plină cu zerouri
matrice_clasica = []
for i in range(linii):
    rand = []
    for j in range(coloane):
        rand.append(0)
    matrice_clasica.append(rand)

# Același lucru, folosind List Comprehension imbricat
matrice_comprehension = [[0 for j in range(coloane)] for i in range(linii)]

print("Matrice 3x4 generată rapid:")
for rand in matrice_comprehension:
    print(rand)

# Matrice generată dinamic pe baza indecșilor (ex: i + j)
matrice_indecsi = [[i + j for j in range(coloane)] for i in range(linii)]
print("\nMatrice 3x4 bazată pe suma indecșilor (i+j):")
for rand in matrice_indecsi:
    print(rand)
```

## 8. Paradigme de programare (Procedurală vs Funcțională) și Iteratori 
### Descriere concept
- **Procedural**: rezolvi problema pas cu pas, cu variabile și bucle.
- **Funcțional**: folosești funcții care transformă datele (`map`, `filter`) și eviți modificări inutile.

Un **iterator** este un obiect care furnizează elemente pe rând.

Bucla `for` folosește iteratori în mod intern.

### Diferențe față de C++
C++ are și el iteratori (`begin`, `end`) și stil procedural.

În Python, utilizarea iteratorilor este mai simplă pentru începători (`for x in lista`).

### Cod și explicații
În cod comparăm abordarea procedurală cu una funcțională și folosim explicit `iter()` + `next()`.

```python
# 8) Procedural vs funcțional + iteratori
numere = [1, 2, 3, 4, 5]

# Procedural: construim lista pas cu pas
patrate_proc = []
for x in numere:
    patrate_proc.append(x * x)
print("Procedural:", patrate_proc)

# Funcțional: map aplică o funcție fiecărui element
patrate_func = list(map(lambda x: x * x, numere))
print("Funcțional (map):", patrate_func)

# Iterator explicit
it = iter(numere)
print("next(it):", next(it))
print("next(it):", next(it))
```

## 8.1 Funcții anonime (lambda), map și filter

### Descriere concept
Până acum am definit funcții folosind cuvântul cheie `def`. Când avem nevoie de o funcție scurtă, cu o singură expresie, pe care o folosim o singură dată (de exemplu, ca parametru pentru altă funcție), putem folosi **funcțiile anonime** (denumite și funcții `lambda`).
- Sintaxa: `lambda parametri: expresie_returnata`

**`map()`** și **`filter()`** sunt funcții care transformă sau filtrează colecții de date. Ele primesc ca prim argument o *altă funcție* (de obicei o funcție lambda) și ca al doilea argument o colecție (listă, tuplu etc.).
- **`map(functie, colectie)`**: aplică funcția pe fiecare element din colecție rând pe rând și returnează rezultatele modificate.
- **`filter(functie, colectie)`**: testează fiecare element cu funcția; păstrează în rezultat doar elementele pentru care funcția a returnat `True`.

*Atenție:* `map` și `filter` returnează un "iterator" (un obiect care generează valorile doar atunci când sunt cerute, pentru a economisi memorie). Pentru a vedea rezultatul direct, trebuie să îl convertim într-o listă folosind `list()`.

### Cod și explicații


```python
# 8.1) Lambda, map și filter

# 1. Lambda vs def clasic
def dubleaza_clasic(x):
    return x * 2

# Aceeași logică, scrisă într-o singură linie
dubleaza_lambda = lambda x: x * 2

print("Clasic:", dubleaza_clasic(5))
print("Lambda:", dubleaza_lambda(5))

# 2. Folosirea map()
numere = [1, 2, 3, 4, 5]

# Vrem să adunăm 10 la fiecare număr
# map va lua lambda x: x + 10 și o va aplica pe 1, apoi pe 2, apoi pe 3...
numere_marite = list(map(lambda x: x + 10, numere))
print("\nRezultat map (adunare +10):", numere_marite)

# 3. Folosirea filter()
# Vrem să păstrăm doar numerele strict mai mari ca 2
numere_filtrate = list(filter(lambda x: x > 2, numere))
print("\nRezultat filter (x > 2):", numere_filtrate)

# 4. Combinarea filter + map
# Vrem să dublăm DOAR numerele impare
# Pas 1: filtrăm imparele
impare = filter(lambda x: x % 2 != 0, numere)
# Pas 2: aplicăm map pentru a le dubla
impare_dublate = list(map(lambda x: x * 2, impare))
print("\nRezultat filter + map (dublează doar imparele):", impare_dublate)
```

### 9.1 Numpy - Puterea calculului vectorial

Numpy ne permite să creăm structuri de date mai eficiente decât listele standard, oferind metode diverse de inițializare și operații matematice directe.


```python
# 9.1) Explorare Numpy 
import numpy as np

# Diferite metode de a crea array-uri
c_zeros = np.zeros(4) # Array cu 4 elemente inițializate cu 0 
a_float = np.array([1, 2, 3, 4], dtype='float') # Convertire listă în float 
matrice_complexa = np.full((3, 3), 6, dtype='complex') # Matrice 3x3 cu nr. complexe 

print("Array de zerouri:", c_zeros)
print("Array float:", a_float)
print("Matrice complexă:\n", matrice_complexa)

# Generare de valori distribuite uniform
# Creează 5 elemente la distanțe egale în intervalul [0, 10]
lin_space = np.linspace(0, 10, 5) 
print("\nLinspace original:", lin_space)

# Operațiile se aplică tuturor elementelor simultan (fără bucle for)
lin_space **= 2 # Ridicarea la pătrat pentru toate elementele 
print("Linspace ridicat la pătrat:", lin_space)

# Putem converti coloane de date în Numpy array-uri pentru viteza de calcul
note_simple = [7, 8, 9, 10]
note_np = np.array(note_simple)
print("Media notelor (via Numpy):", np.mean(note_np))
```

### De ce Numpy e mai rapid? Demonstrație cu `%timeit`

Numpy nu execută bucla în Python – o rulează în C, intern. Pe liste mari diferența de viteză devine dramatică.

```python
import numpy as np

n = 1_000_000
lista_python = list(range(n))
array_numpy = np.arange(n, dtype=float)

print("Buclă Python (sumă de pătrate):")
%timeit sum(x * x for x in lista_python)

print("Numpy vectorizat (sumă de pătrate):")
%timeit np.sum(array_numpy ** 2)
```

### 9.2 Pandas - Procesarea fișierului "housing.csv"

Pandas transformă fișierele de date în obiecte de tip `DataFrame` (tabele). Vom folosi setul de date `housing`, care conține informații despre imobiliare (populație, număr de camere, valoarea medie a caselor etc.).

```python
# 9.2) Analiză avansată cu Pandas
import pandas as pd

# 1. Încărcarea datelor
# ATENȚIE: Fișierul "housing.csv" trebuie să fie în același folder!
df = pd.read_csv("housing.csv")

# 2. Vizualizarea și explorarea rapidă
print("Primele 3 rânduri din tabel:")
print(df.head(3)) 

print("\nStatistici descriptive (număr, medie, deviație standard, min, max):")
# Selectăm doar câteva coloane pentru a nu aglomera ecranul
print(df[["housing_median_age", "total_rooms", "median_house_value"]].describe())

# 3. Operații pe coloane și filtrări
# Adăugăm o coloană nouă calculată: Numărul mediu de camere per locuință
df["camere_per_locuinta"] = (df["total_rooms"] / df["households"]).round(2)

# Filtrăm zonele cu case vechi (peste 40 de ani) dar cu valoare sub 250.000
case_vechi_ieftine = df[(df["housing_median_age"] > 40) & (df["median_house_value"] < 250000)]
print("\nZone cu case vechi (>40 ani) și ieftine (<250k):")
print(case_vechi_ieftine[["housing_median_age", "median_house_value", "ocean_proximity"]].head())

# 4. Grupări complexe
# Media valorii caselor și totalul populației pentru fiecare zonă (ocean_proximity)
analiza_zone = df.groupby("ocean_proximity").agg({
    "median_house_value": "mean",
    "population": "sum"
}).reset_index()

print("\nAnaliza pe zone (Valoare medie și Populație totală):")
print(analiza_zone)
```

### 9.3 Matplotlib - Vizualizarea rezultatelor

Vom folosi Matplotlib pentru a descoperi vizual relațiile din date. De exemplu, există o corelație între venitul mediu dintr-o zonă (`median_income`) și valoarea caselor (`median_house_value`)?

```python
# 9.3) Vizualizare date imobiliare cu Matplotlib
import matplotlib.pyplot as plt

# Setăm dimensiunea figurii
plt.figure(figsize=(9, 5))

# Creăm un grafic cu puncte (scatter plot)
# X = Venitul mediu (median_income)
# Y = Valoarea casei (median_house_value)
plt.scatter(
    df["median_income"], 
    df["median_house_value"], 
    alpha=0.6,          # Grad de transparență pentru a vedea punctele suprapuse
    color="steelblue",  # Culoarea punctelor
    s=20                # Dimensiunea punctelor
)

# Configurăm titlul și etichetele axelor
plt.title("Corelația între Venitul Mediu și Valoarea Caselor")
plt.xlabel("Venit Mediu (median_income)")
plt.ylabel("Valoarea Medie a Casei (median_house_value)")

# Adăugăm caroiaj
plt.grid(True, linestyle="--", alpha=0.5)

# Afișăm graficul pe ecran
plt.show()
```
